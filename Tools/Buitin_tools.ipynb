## TavilySearch


```python
from langchain_tavily import TavilySearch
from dotenv import load_dotenv

load_dotenv()

tool = TavilySearch(
	max_results=5,
	topic="general",
)

res = tool.invoke({'query' : "what is recent news ?"})


for result in res['results']:
	print(result['title'])
	print(result['url'])
	print(result['content'])
	print("------")
```

## Building Tool Using Structured Tool


```python
from pydantic import BaseModel, Field
from typing import Literal
from langchain_core.tools import StructuredTool
```

```python
class WeatherInput(BaseModel):
    location: str = Field(description="Name of the city")
    units : Literal["celsius", "fahrenheit"] = Field(default="celsius", description="Temperature unit preference")
```

```python
def get_whether(location: str, units: str) -> str:
	"""Get current weather and optional forecast."""
	temp = 22 if units == "celsius" else 72
	result = f"Current weather in {location}: {temp} degrees {units[0].upper()}"
	return result
```

```python
whether_tool = StructuredTool.from_function(
	func=get_whether,
	name="get_whether",
	description="Get current weather and optional forecast.",
	args_schema=WeatherInput
)
```

```python
res = whether_tool.invoke({"location": "San Francisco", "units": "fahrenheit"})
print(res)
```

## ToolKit


```python
from langchain.tools import tool
```

```python
@tool
def add(x: float, y: float) -> float:
	"""Add x and y"""
	return x + y

@tool
def multiply(x: float, y: float) -> float:
	"""Multiply x and y"""
	return x * y
```

```python
class Math_toolkit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = Math_toolkit()
tools = toolkit.get_tools()

for tool in tools:
	print(tool.name, "->", tool.description)
```
