```python
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain.agents import create_agent
import requests

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model='gemini-2.0-flash')
```

```python
@tool
def get_wheather(city: str):
    """Get current wheather information for a given city."""
    response = requests.get(f'https://wttr.in/{city}?format=3')
    return response.text
```

```python
@tool
def search_flights(source: str, destination: str):
    """Find top 5 flights between two cities."""

    res = llm.invoke(f"Give me top 5 flights list from {source} to {destination} with only flight names and prices only.")
    return f"Flights from {source} to {destination}: {res}"
```

```python
@tool
def suggest_itinerary(city: str):
	"""Suggest top 3 attractions in a given city use attractions dictonary as refrence."""
	attractions = {
		'Paris' : ['Eiffel Tower', 'Louvre Museum', 'Arc de Triomphe'], 
		'New york' : ['Statue of Liberty', 'Empire State Building', 'Times Square'],
		'Tokyo' : ['Tokyo Tower', 'Shinjuku Gyoen National Garden', 'Senso-ji Temple']
	}
	return attractions.get(city, ['No attractions found for this city.'])
```

```python
agent = create_agent(
	llm,
	tools=[get_wheather, search_flights, suggest_itinerary],
	system_prompt="You are an AI travel agent. You will be given a destination, and you will provide a trip plan."
)
```

```python
res = agent.invoke({"messages": [{"role": "user", "content": "Plan a 3-day trip from Delhi to Paris. Check weather and flights too."}]})
```
